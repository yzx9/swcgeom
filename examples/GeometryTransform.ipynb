```python
import math

import matplotlib.pyplot as plt
from swcgeom import Tree
from swcgeom.analysis import draw
from swcgeom.transforms import Translate, Scale, RotateZ
```

```python
t = Tree.from_eswc("./data/toydata.swc")

fig, axes = plt.subplots(1, 4)
draw(t, ax=axes[0])
draw(Translate(10,10,10)(t), ax=axes[1])
draw(Scale(10,10,10)(t), ax=axes[2])
draw(RotateZ(0.5 * math.pi)(t), ax=axes[3])
fig.tight_layout()
```
